```python
def horizontal_merge(df1, df2):
    return df1.merge(df2, left_on = ['country', 'Year'], right_on = ['country', 'Year'])
```

```python
def vertical_merge(df1, df2):
    return pd.concat([df1, df2], ignore_index = True)
```

```python
def indicator_melt(df, indicator):
    return pd.melt(df, id_vars = df.columns[0], value_vars = df.columns[1:], var_name = 'Year', value_name = indicator)
```

```python
def pop_melt(df, indicator):
    return pd.melt(df, id_vars = df.columns[:2], value_vars = df.columns[2:], var_name = 'Year', value_name = indicator)
```

```python
male_life_melt = indicator_melt(male_life, 'LifeExpectancy')
female_life_melt= indicator_melt(female_life, 'LifeExpectancy')
```

```python
male_death_melt = pop_melt(male_lung_death, 'Death')
female_death_melt= pop_melt(female_lung_death, 'Death')
```

```python
male = horizontal_merge(male_life_melt,male_death_melt)
female = horizontal_merge(female_life_melt,female_death_melt)
```

```python
test = vertical_merge(female, male)
```
